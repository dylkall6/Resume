```python
import pandas as pd

data1 = pd.read_excel("644_Data_1.xlsx", index_col = [1])
```

```python
data1.shape
```

```python
data1.keys()
```

```python
data1
```

```python
data = data1

features = data1.keys()

df = pd.DataFrame(data, columns = features)

print(df.shape)

print(features)
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

plt.figure(figsize=(11,11))

df_large = df.iloc[:,:13]

correlation_mat = df_large.corr()

sns.heatmap(correlation_mat, annot = True)

plt.title("Correlation matrix of Data1 (Q4 2017 to Q3 2018)")

plt.xlabel("Variables")

plt.ylabel("Variables")

plt.show()
```
